```python
# Loading and Inspecting Data

# Load various data formats (CSV, Excel) into Pandas DataFrames.
import pandas as pd
df=pd.read_csv('loan_data_train.csv')
print(df.head())

# Explore DataFrame attributes like shape, columns, dtypes, head, tail, info, describe.

print(df.shape)
print(df.columns)        
print(df.dtypes)         
print(df.head(3))        
print(df.tail(2))        
df.info()                
print(df.describe())

# Practice selecting columns and rows using different methods (indexing, slicing, loc, iloc).

# Selecting columns
print(df['Amount.Requested'])        
print(df[['Amount.Requested', 'Interest.Rate']]) 

# Selecting rows
print(df.iloc[0])        
print(df.loc[0:2])       

# Selecting rows and columns together
print(df.iloc[0:3, 1:3]) 
print(df.loc[0:3, ['Amount.Requested', 'Interest.Rate']]) 
```

```python
# Q2. Data Cleaning and Preparation

# Identify missing values using isnull and isna.
miss = df.isnull()
missing = df.isnull().sum()

# Handle missing values using fillna, dropna, interpolation.
df['Amount.Requested'].fillna(df['Amount.Requested'].mean(), inplace=True)

df['Amount.Requested'].dropna()
#df.dropna(axis=1)
#df.dropna(how='all')

df.interpolate()
df.interpolate(method='polynomial', order=2)

# Apply scaling techniques (min-max, z-score) to numerical columns.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['scaled_column'] = scaler.fit_transform(df[['Amount.Requested']])

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['z_scaled_column'] = scaler.fit_transform(df[['Amount.Requested']])

# Create dummy variables for categorical columns.
df = pd.get_dummies(df, columns=['Amount.Requested'])

print(df)
```
