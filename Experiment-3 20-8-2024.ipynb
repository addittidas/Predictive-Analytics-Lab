```python
# Q2. Data Loading and Preprocessing:

# Load a dataset (e.g., CSV, Excel) using NumPy.
import numpy as np

# Load a CSV file
data = np.genfromtxt('loan_data_train.csv', delimiter=',', skip_header=1)
print(data)

# Clean and preprocess the data (handling missing values, normalization, standardization).
missing_values = np.isnan(data)
col_mean = np.nanmean(data, axis=0)
indices = np.where(np.isnan(data))
data[indices] = np.take(col_mean, indices[1])

print("\nNormalization\n")
data_min = np.min(data, axis=0)
data_max = np.max(data, axis=0)
data_normalized = (data - data_min) / (data_max - data_min)
print(data_normalized)

print("\nStandardization\n")
data_mean = np.mean(data, axis=0)
data_std = np.std(data, axis=0)
data_standardized = (data - data_mean) / data_std

#Calculate mean, median, standard deviation, variance, and other statistical measures.
mean = np.mean(data, axis=0)  # Mean of each column
median = np.median(data, axis=0)  # Median of each column
std_dev = np.std(data, axis=0)  # Standard deviation of each column
variance = np.var(data, axis=0)  # Variance of each column

min_values = np.min(data, axis=0)
max_values = np.max(data, axis=0)
sum_values = np.sum(data, axis=0)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
print("Variance:", variance)
print("Min:", min_values)
print("Max:", max_values)
print("Sum:", sum_values)
```
